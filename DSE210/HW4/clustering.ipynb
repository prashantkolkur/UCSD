```python
import numpy as np
import pandas as pd
from sklearn.cluster import KMeans
```

```python
classes = pd.read_csv('classes.txt',delim_whitespace=True,names=['id','animal'])
```

```python
animal_features = open('predicate-matrix-continuous.txt', 'r')
```

```python
a = animal_features.readlines()
```

```python
b = []
for i in range(len(a)):
    b.append(list(np.array(np.matrix(a[i]))[0]))
    
```

```python
kmeans = KMeans(n_clusters=10, random_state=0).fit(b)
```

```python
classes
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

```python
classes['cluster_label'] = kmeans.labels_
```

```python
classes
```

```python
for i in range(10):
    print 'Cluster: ',i
    print classes[classes['cluster_label']==i]['animal'],'\n'
```

```python
from scipy.cluster.hierarchy import dendrogram, linkage
#from pylab import rcParams
from matplotlib import pyplot as plt
```

```python
h = linkage(b, 'ward')
fig = plt.figure(figsize=(10, 15))
dn = dendrogram(h, orientation='right', labels=list(classes['animal']))
#rcParams['figure.figsize'] = 5,10
plt.show()
```
