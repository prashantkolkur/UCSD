```python
import urllib2
import pandas as pd
import numpy as np
import re

urls = ['https://www2.census.gov/programs-surveys/saipe/datasets/2001/2001-state-and-county/est01all.dat',
        'https://www2.census.gov/programs-surveys/saipe/datasets/2002/2002-state-and-county/est02all.dat',
        'https://www2.census.gov/programs-surveys/saipe/datasets/2003/2003-state-and-county/est03all.dat',
        'https://www2.census.gov/programs-surveys/saipe/datasets/2004/2004-state-and-county/est04all.txt',
        'https://www2.census.gov/programs-surveys/saipe/datasets/2005/2005-state-and-county/est05all.txt',
        'https://www2.census.gov/programs-surveys/saipe/datasets/2006/2006-state-and-county/est06all.txt',
        'https://www2.census.gov/programs-surveys/saipe/datasets/2007/2007-state-and-county/est07all.txt',
        'https://www2.census.gov/programs-surveys/saipe/datasets/2008/2008-state-and-county/est08all.txt',
        'https://www2.census.gov/programs-surveys/saipe/datasets/2009/2009-state-and-county/est09all.txt',
        'https://www2.census.gov/programs-surveys/saipe/datasets/2010/2010-state-and-county/est10all.txt',
        'https://www2.census.gov/programs-surveys/saipe/datasets/2011/2011-state-and-county/est11all.txt',
        'https://www2.census.gov/programs-surveys/saipe/datasets/2012/2012-state-and-county/est12all.txt',
        'https://www2.census.gov/programs-surveys/saipe/datasets/2013/2013-state-and-county/est13all.txt',
        'https://www2.census.gov/programs-surveys/saipe/datasets/2014/2014-state-and-county/est14all.txt',
        'https://www2.census.gov/programs-surveys/saipe/datasets/2015/2015-state-and-county/est15all.txt']

def getUrl(urls):
    cols=['Year','State FIPS', 'County FIPS','Poverty Estimate All Ages', 'Poverty Percent All Ages', 'Poverty Estimate Under Age 18', 'Poverty Percent Under Age 18', 'Poverty Estimate Ages 5-17', 'Poverty Percent Ages 5-17', 'Median Household Income','Name','Postal']
    df = pd.DataFrame(columns=cols)
    to_numeric_cols = ['Year','Poverty Estimate All Ages', 'Poverty Percent All Ages', 'Poverty Estimate Under Age 18', 'Poverty Percent Under Age 18', 'Poverty Estimate Ages 5-17', 'Poverty Percent Ages 5-17', 'Median Household Income']
    #to_int_cols = ['Year','Poverty Estimate All Ages', 'Poverty Estimate Under Age 18', 'Poverty Estimate Ages 5-17', 'Median Household Income']
    #to_float_cols = ['Poverty Percent All Ages', 'Poverty Percent Under Age 18', 'Poverty Percent Ages 5-17']
    for url in urls:
        response = urllib2.urlopen(url)
        lines = response.read().split('\n')
        del lines[-1]
        l = []
        for line in lines:
            l.append([re.findall(r'20\d*',url)[0],line[0:2].strip(),line[3:6].strip(),line[7:15].strip(),line[34:38].strip(),line[49:57].strip(),line[76:80].strip(),line[91:99].strip(),line[118:122].strip(),line[133:139].strip(),line[193:238].strip(),line[239:241].strip()])
        df1 = pd.DataFrame(l ,columns=cols)
        df1[to_numeric_cols] = df1[to_numeric_cols].apply(pd.to_numeric, errors='coerce')
        #df1[to_int_cols] = df1[to_int_cols].astype(int)
        #df1[to_float_cols] = df1[to_float_cols].astype(float)
        df = df.append(df1, ignore_index=True)
    return df

df = getUrl(urls)
```

```python
import urllib2
import xlrd
import pandas as pd
import numpy as np
import re

urls = ['https://www2.census.gov/programs-surveys/school-finances/tables/2001/secondary-education-finance/elsec01t.xls',
        'https://www2.census.gov/programs-surveys/school-finances/tables/2002/secondary-education-finance/elsec02t.xls',
        'https://www2.census.gov/programs-surveys/school-finances/tables/2003/secondary-education-finance/elsec03t.xls',
        'https://www2.census.gov/programs-surveys/school-finances/tables/2004/secondary-education-finance/elsec04t.xls',
        'https://www2.census.gov/programs-surveys/school-finances/tables/2005/secondary-education-finance/elsec05t.xls',
        'https://www2.census.gov/programs-surveys/school-finances/tables/2006/secondary-education-finance/elsec06t.xls',
        'https://www2.census.gov/programs-surveys/school-finances/tables/2007/secondary-education-finance/elsec07t.xls',
        'https://www2.census.gov/programs-surveys/school-finances/tables/2008/secondary-education-finance/elsec08t.xls',
        'https://www2.census.gov/programs-surveys/school-finances/tables/2009/secondary-education-finance/elsec09t.xls',
        'https://www2.census.gov/programs-surveys/school-finances/tables/2010/secondary-education-finance/elsec10t.xls',
        'https://www2.census.gov/programs-surveys/school-finances/tables/2011/secondary-education-finance/elsec11t.xls',
        'https://www2.census.gov/programs-surveys/school-finances/tables/2012/secondary-education-finance/elsec12t.xls',
        'https://www2.census.gov/programs-surveys/school-finances/tables/2013/secondary-education-finance/elsec13t.xls',
        'https://www2.census.gov/programs-surveys/school-finances/tables/2014/secondary-education-finance/elsec14t.xls',
        'https://www2.census.gov/programs-surveys/school-finances/tables/2015/secondary-education-finance/elsec15t.xls',
        ]

pea = pd.DataFrame(columns=['Year','FIPS', 'TOTALEXP', 'ENROLL'])
for url in urls:
    File = urllib2.urlopen(url)
    Book = pd.ExcelFile(File)
    df1 = Book.parse()
    year = int(re.findall(r'20\d*',url)[0])
    if year == 2001:
        df1 = df1[['FIPS', 'TOTALEXP', 'ENROLL']]
        df1.FIPS = df1.FIPS.apply(lambda x: x if x/100000 == 0 else x/10000)
    elif year == 2002:
        df1 = df1[['FIPS', 'TOTALEXP', 'ENROLL']]
        df1.FIPS = df1.FIPS.apply(lambda x: x/1000000)
    elif year == 2007:
        df1 = df1[['CONUM', 'TOTALEXP', 'ENROLL']]
        df1 = df1.rename(columns={'CONUM':'FIPS'})
        df1.FIPS = df1.FIPS.apply(lambda x: int(x) if ((type(x)!=int) & (x.strip()!='')) else 0)
    else:
        df1 = df1[['CONUM', 'TOTALEXP', 'ENROLL']]
        df1 = df1.rename(columns={'CONUM':'FIPS'})
    df1['Year'] = year
    pea = pea.append(df1, ignore_index=True)
```

```python
pea_s1 = pea.groupby(['Year','FIPS'], as_index=False).sum()
pea_s1['Per Student TOTALEXP'] = pea_s1.TOTALEXP/pea_s1.ENROLL
merged_data = pd.merge(county_stat[['Year', 'FIPS', 'Postal', 'Name', 'Poverty Percent All Ages']], pea_s1, on=['Year', 'FIPS'])
merged_data = merged_data.replace(np.inf, np.nan).dropna(axis=0)
```

```python
cols = ['FIPS', 'Postal', 'Name', 'Per_Student_TOTALEXP', 'Var of Poverty Percent All Ages','Var of Per Student TOTALEXP', 'Correlation of PPAA and Per_Student_TOTALEXP']

PEA_list = []
for fips in merged_data.FIPS.unique():
    df = merged_data[merged_data.FIPS == fips]
    a1 = df['Poverty Percent All Ages'].iloc[-6:-1].reset_index()
    a2 = df['Per Student TOTALEXP'].iloc[0:5].reset_index()
    corr = a1['Poverty Percent All Ages'].corr(a2['Per Student TOTALEXP'])
    var_a1 = a1['Poverty Percent All Ages'].cov(a1['Poverty Percent All Ages'])
    #corr = df[['Poverty Percent All Ages', 'Per Student TOTALEXP']].corr()
    var_a2 = a2['Per Student TOTALEXP'].cov(a2['Per Student TOTALEXP'])
    l = [fips, df['Postal'].iloc[0], df['Name'].iloc[0], a2['Per Student TOTALEXP'].mean(), var_a1, var_a2, corr]
    PEA_list.append(l)
PEA = pd.DataFrame(PEA_list, columns=cols)
PEA = PEA.dropna()
```
