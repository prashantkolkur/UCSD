(6) calculate the average word length for the text

> TEXT =
> On a dark desert highway, cool wind in my hair
> Warm smell of colitas, rising up through the air
> Up ahead in the distance, I saw a shimmering light
> My head grew heavy and my sight grew dim
> I had to stop for the night
> There she stood in the doorway;
> I heard the mission bell
> And I was thinking to myself,
> "This could be Heaven or this could be Hell"
> Then she lit up a candle and she showed me the way
> 

Average word length is the sum of all the lengths of the word tokens in the text, divided by the number of word tokens


```python
TEXT = """On a dark desert highway, cool wind in my hair Warm smell of colitas, rising up through the air Up ahead in the distance, I saw a 
shimmering light My head grew heavy and my sight grew dim I had to stop for the night There she stood in the doorway; I heard the mission bell
And I was thinking to myself, "This could be Heaven or this could be Hell" Then she lit up a candle and she showed me the way
"""
#remove all special characters
for c in ',-;."\n':
    TEXT = TEXT.replace(c, ' ')
#split the text
T = TEXT.split()

#sum of all words
s = 0
for i in range(len(T)):
    s = s + len(T[i])

#Average word length
Avg = (s+0.0)/len(T)
print Avg
```

(13) Write a program to make a list of all prime numbers less than 3000. A prime number can only be divided evenly by 1 and itself.

```python
#List of all prime numbers less than 3000
prime = []
for i in range(2,3000,1):
    j = i/2
    for k in range(2,j+2,1):
        if i == 2:
            break
        elif i%k ==0:
            break
    if k == j+1:
        prime.append(i)
print prime
```
